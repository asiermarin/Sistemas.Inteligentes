```python
vehiculo_2 = vehiculo.drop('class', axis=1)
```

```python
from scipy.stats import zscore

vehiculo_z = vehiculo_2.apply(zscore)
vehiculo_z.head()
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

standar_scaler = StandardScaler()
X_standarD = standar_scaler.fit_transform(vehiculo_2)
```

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist

clusters = range(1, 10)
meanDistorsions = []

for k in clusters:
    model = KMeans(n_clusters = k)
    print(model)
    model.fit(X_standarD)
    prediction = model.predict(X_standarD)
    meanDistorsions.append(
        sum(np.min(
        cdist(X_standarD, model.cluster_centers_, 'euclidean'), axis = 1))/X_standarD.shape[0])
```

```python
plt.plot(clusters, meanDistorsions, 'bx-')
plt.xlabel('A')
plt.xlabel('Avg distorsion')
plt.title('Seleccionar valor de K * Estrategia')
```

```python
kmeans = KMeans(n_clusters = 3, n_init = 15, random_state = 2345)
kmeans.fit(X_standarD)

centroids = kmeans.cluster_centers_

centroids
```

```python
centroids_df = pd.DataFrame(centroids, columns=list(X_standarD))

dataframe_labels = pd.DataFrame(kmeans.labels_, columns = list(['labels']))

dataframe_labels['labels'] = dataframe_labels['labels'].astype('category')

dataframe_labeled = vehiculo_2.join(dataframe_labels)
```

```python
dataframe_analysis = (dataframe_labeled.groupby(['labels'], axis = 0)).head(1234)
dataframe_analysis
```
