```python
#Observamos los datos que tenemos en el dataset y los intentamos entender

#Importamos Pandas
import pandas as pd

#Leemos e importamos los datos/registros del csv
df = pd.read_csv('IPL_Ball-by-Ball_2008-2020.csv')

#Mostramos los 20 primeros registros que vienen en el csv
df.head(20)


#Observamos que los datos que disponemos se tratan de datos de partidos de criquet disputados por diferentes equipos
#Cada fila de datos hace referencia a los datos correspondientes a los equipos que juegan el partido
#Se observa que cada partido tiene un ID comun.
```

```python
#Contamos el numero de datos por cada variable para ver si hay datos sesgados
df.count()
```

```python
#Realizamos una limpieza de datos
#Eliminamos las variables que no son significativas

#Eliminamos 'id', 'non_striker', 'bowler' ya que no aportan valor a la prediccion que queremos realizar
#Eliminamos 'dimissal_kind', ,'player_dismissed','fielder','extras_type' ya que contienen valores nulos 
df = df.drop(columns=['id', 'dismissal_kind', 'player_dismissed', 'fielder', 'extras_type', 'batsman', 'non_striker', 'bowler'],axis=1)

#Eliminamos los datos nulos  
df = df.dropna(how='any')

#Después, mostramos los 20 primeros registros actualizados
df.head(20)
```

```python
#Contamos de nuevo el numero de datos para ver si los datos han sido filtrados correctamente
df.count()
```

```python
# Importamos Preprocessing
from sklearn import preprocessing

#Usamos el hot encoder ya que sirve para variable numericas y categoricas
#Vamos a transformar las variables categoricas 'batting_team' y 'bowling_team' en variables numericas
#Se van a convertir dichas variables categoricas en columnas con 0 y 1

# Cargamos el one-hot encoder
hot_encoder=preprocessing.OneHotEncoder(sparse=False)

#Vamos a generar un nuevo DataFrame formado por los datos a modificar para realizar las operaciones
# Aplicamos el encoder las columnas 'batting_team','bowling_team'
datos_equipos = pd.DataFrame(hot_encoder.fit_transform(df[['batting_team','bowling_team']]))

#Nombramos las nuevas columnas
datos_equipos.columns = hot_encoder.get_feature_names(['batting_team','bowling_team'])

# Eliminamos las variables 'batting_team' y 'bowling_team de nuestro dataset original
df = df.drop(columns=['batting_team','bowling_team'], axis =1)

#Añadimos el nuevo DataFrame al viejo
df = pd.concat([df,teams_data], axis=1)

#Realizamos de nuevo un eliminado de datos nulos del nuevo dataframe 
df = df.dropna(how='any')

#Mostramos los 20 primeros registros que vienen en el nuevo csv
df.head(20)
```

```python
#Interrelacionamos dos variables de los datos del dataset ya filtrado correctamente.
#En nuestro caso vamos a usar la Regresion Lineal.
#Dividimos los datos que tenemos en datos de entrenamiento y de testeo
#80% de los datos para entrenamiento y un 20% para tests.

#Las variables que vamos a relacionar en nuestro caso son "over" y "batsman_runs"
#1 Over se trata de 6 lanzamientos legales por parte del lanzador de la bola al bateador
#Total_runs se trata de el numero de carreras totales realizadas por un jugador, es decir, los puntos totales

#Importamos la funcion de entrenamiento
from sklearn.model_selection import train_test_split

#Definimos la variable x (independiente)
X = df[['over']]

#Definimos la variable y
y = df[['batsman_runs']]

#Definimos que los datos de testeo van a ser de un tamaño del 20%
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20)


from sklearn import linear_model

# Inicializamos el modelo de regresión lineal
regression = linear_model.LinearRegression()

#Generamos los valores de entrenamiento
regression_train_x = X_train.values
regression_train_y = y_train.values

# Entrenamos el modelo de datos
regression = regression.fit(regression_train_x.reshape(-1, 1), regression_train_y.reshape(-1, 1))
```

```python
#Visualizamos los resultados y imprimimos los coeficientes de la regresión, la media de los mínimos cuadrados y el R^2 

#Importamos las librearias necesarias
from sklearn import metrics
import matplotlib.pyplot as plt

#Generamos los valores de testeo
regression_test_x = X_test.values
regression_test_y = y_test.values

#Realizamos la prediccion
predictions = regression.predict(regression_test_x.reshape(-1, 1))

# Muestra el coeficiente la función
print(regression.coef_)

# Muestra la perdida de regresion del error cuadratico medio
print(metrics.mean_squared_error(regression_test_y, predictions))

# Puntuamos la regresion como criterio de evaluacion para R^2 (la mejor puntuacion posible es 1.0)
print(regression.score(X, y))

#Mostramos el resultado en una gráfica
plt.scatter(regression_test_x, regression_test_y,  color='black')
plt.plot(regression_test_x, predictions, color='red', linewidth=3)
plt.show()

#Observamos que el porcentaje obtenido es muy bajo por lo que vamos a intentar realizar una mejora.
#Para ello, modificaremos primeramente el porcentaje de datos de testeo y en el caso de no tener una mejora 
#modificaremos la variable independiente
```

```python
#Hemos decidido aumentar el numero de datos de testeo
#Hemos pasado de coger el 80% de datos de entrenamiento y el 20% de testeo
#a coger 60% de entrenamiento y 40% de testeo

#Importamos la funcion de entrenamiento
from sklearn.model_selection import train_test_split

#Definimos la variable x (independiente)
X = df[['over']]

#Definimos la variable y
y = df[['batsman_runs']]

#Realizamos la modificación y definimos que los datos de testeo van a ser de un tamaño del 40%
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.40)


from sklearn import linear_model

# Inicializamos el modelo de regresión lineal
regression = linear_model.LinearRegression()

#Generamos los valores de entrenamiento
regression_train_x = X_train.values
regression_train_y = y_train.values

# Entrenamos el modelo de datos
regression = regression.fit(regression_train_x.reshape(-1, 1), regression_train_y.reshape(-1, 1))

#Importamos las librearias necesarias
from sklearn import metrics
import matplotlib.pyplot as plt

#Generamos los valores de testeo
regression_test_x = X_test.values
regression_test_y = y_test.values

#Realizamos la prediccion
predictions = regression.predict(regression_test_x.reshape(-1, 1))

# Muestra el coeficiente la función
print(regression.coef_)

# Muestra la perdida de regresion del error cuadratico medio
print(metrics.mean_squared_error(regression_test_y, predictions))

# Puntuamos la regresion como criterio de evaluacion para R^2 (la mejor puntuacion posible es 1.0)
print(regression.score(X, y))

#Mostramos el resultado en una gráfica
plt.scatter(regression_test_x, regression_test_y,  color='black')
plt.plot(regression_test_x, predictions, color='red', linewidth=3)
plt.show()

#Modificando el porcentaje de numero de datos de entrenamiento que se va a usar, 
#se observa que el procentaje no mejora.
#Sigue siendo un 0,76% practicamente, por lo que vamos a modificar la variable independiente para intentar
#mejorar el porcentaje
```

```python
#Para la eleccion de una variable que pueda mejorar nuestro algoritmo, vamos a predecir las dos variables
#mas significativas de nuestro dataset mediante el SelectBest utilizado en la Practica 1
#Cabe destacar que podemos realizar esta prediccion de la variable mas significativa ya que todas las
#variables del nuevo dataset son numericas.

#Importamos las librerias
from sklearn.feature_selection import SelectKBest, chi2

#Indicamos la variable sobre la que queremos que nos haga la relación
X = df.loc[:,df.columns!='batsman_runs']
y = df[['batsman_runs']]
selector = SelectKBest(chi2, k=2)
selector.fit(X, y)

#Nos devuelve las variables mas significativas respecto a la variable indicada
print(X.columns[selector.get_support(indices=True)])

#Observamos que una de las variables que nos imprime es la que ya hemos utilizado anteriormente para
#relacionarla con 'batsman_runs'
#Por lo que elegimos 'total_runs' como nueva variable independiente
```

```python
#Hemos obtenido una precision del 0,07% por lo que intentamos mejorar los resultados
#cambiando la variable independiente.
#Ahora en vez de tratar con los datos de 'over', trataremos con los de 'total_runs'

#Importamos la funcion de entrenamiento
from sklearn.model_selection import train_test_split

#Definimos la variable x (independiente)
X = df[['total_runs']]

#Definimos la variable y
y = df[['batsman_runs']]

#Realizamos la modificación y definimos que los datos de testeo van a ser de un tamaño del 40%
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.40)


from sklearn import linear_model

# Inicializamos el modelo de regresión lineal
regression = linear_model.LinearRegression()

#Generamos los valores de entrenamiento
regression_train_x = X_train.values
regression_train_y = y_train.values

# Entrenamos el modelo de datos
regression = regression.fit(regression_train_x.reshape(-1, 1), regression_train_y.reshape(-1, 1))

#Importamos las librearias necesarias
from sklearn import metrics
import matplotlib.pyplot as plt

#Generamos los valores de testeo
regression_test_x = X_test.values
regression_test_y = y_test.values

#Realizamos la prediccion
predictions = regression.predict(regression_test_x.reshape(-1, 1))

# Muestra el coeficiente la función
print(regression.coef_)

# Muestra la perdida de regresion del error cuadratico medio
print(metrics.mean_squared_error(regression_test_y, predictions))

# Puntuamos la regresion como criterio de evaluacion para R^2 (la mejor puntuacion posible es 1.0)
print(regression.score(X, y))

#Mostramos el resultado en una gráfica
plt.scatter(regression_test_x, regression_test_y,  color='black')
plt.plot(regression_test_x, predictions, color='red', linewidth=3)
plt.show()


#Observamos que con el cambio de la variable independiente por la que hemos obtenido con el
#SelectBest, hay una gran mejoria de la precision.
#En este caso pasamos de tener un 0,7% de precision a un 95%
#Es obvia la elevada precision ya que el numero total de juegos depende de los juegos que realicen los
#jugadores del partido
```
